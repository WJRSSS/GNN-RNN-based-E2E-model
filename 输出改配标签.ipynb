```python
#提取target数据
time_range = pd.DatetimeIndex(data_df.index)
series_dict_target = {}
#series_dict[0] = data_df['value']
i = 0
for col in data_df.columns:
    if i < ((target_width - 1) * target_width):
        series_dict_target[i] = data_df[col]
    i += 1
    if i > 64: break
target_df = pd.DataFrame(index = time_range,data=series_dict_target)
```

```python
refining = []
refined = []
```

```python
column_names = np.array(data_df.columns[:target_width * (target_width - 1)])
```

```python
for i in range(target_width * (target_width - 1)):
    if(column_names[i].split('_')[0] not in refining):
        refining.append(column_names[i].split('_')[0])
    refined.append(column_names[i].split('_')[2])
```
